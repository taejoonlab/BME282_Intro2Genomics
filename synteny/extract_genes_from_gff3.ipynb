```python
# Download GFF3 file from EnsEMBL
# You can modify this if you want another GFF3.

url_gff3 = "https://ftp.ensembl.org/pub/release-115/gff3/homo_sapiens/Homo_sapiens.GRCh38.115.chromosome.22.gff3.gz"

!wget -O human_chr22.gff3.gz {url_gff3}
!gunzip -d human_chr22.gff3.gz
```

```python
# check the file
!ls
```

```python
# check the file format
! head human_chr22.gff3
```

```python
filename_gff3 = "human_chr22.gff3"

f_gff = open(filename_gff3, 'r')
f_out = open("%s.genes.txt" % filename_gff3, 'w')
for line in f_gff:
  if line.startswith('#'):
    continue
  tokens = line.strip().split("\t")
  if tokens[2] == "gene":
    f_out.write("%s\n" % line.strip())
    print(line.strip())
f_gff.close()
f_out.close()
```
